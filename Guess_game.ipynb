```python
Secret_ans = "giraffe"
Answer = ""
Answer_count = 0
Answer_limit = 3
Out_of_answer = False

while Answer != Secret_ans and not(Out_of_answer):
    print("Who is a tall animal? ")
    if Answer_count < Answer_limit:
        Answer = input("Enter Answer: ")
        Answer_count += 1
    else:
        Out_of_answer = True

if Out_of_answer :
  print("No more random answers, You Lose! ")
else :
  print("You Win!")
```
